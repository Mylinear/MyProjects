```python
def show_nulls(data):
    '''
    This function plots missing values for each column by observation in the dataset.
    '''
    plt.figure(figsize=(10, 5))
    sns.displot(data=data.isnull().melt(value_name="missing"),
                y="variable",
                hue="missing",
                multiple="fill",
                height=9.6)
    plt.axvline(0.9, color="r")
    plt.show()
```

```python
benzine = ["Gasoline", "Super 95", "Regular","Super E10 95","Super Plus 98","Super Plus E10 98", "Others"]
diesel = ["Biodiesel","Diesel"]
lpg = ["Biogas", "Domestic gas H", "LPG","Liquid petroleum gas", "CNG"]
hybrid = ['Electric/Gasoline']
electric = ['Electric']
def fueltype(x):
    if any (i in x for i in hybrid):
        return 'Hybrid'
    elif any(i in x for i in electric):
        return "Electric"
    elif any(i in x for i in lpg):
        return "LPG/CNG"
    elif any(i in x for i in benzine):
        return "Benzine"
    elif any(i in x for i in diesel):
        return "Diesel"
    else:
        return x
df["fuel"].apply(fueltype).value_counts()
```

```python
df["fuel"] = df["fuel"].apply(fueltype)
df["fuel"]
```

#### consumption

```python
df.consumption.sample(20)
```

```python
def fix(x):
    
    if type(x) == list:
        yeni = ""
        
        for i in x :
            if type(i) == list :
                for j in i :
                    yeni += ", " + j.strip("\n")
            else :        
                yeni += ", " + i.strip("\n")
        return yeni.lstrip(" ,") 

    elif ((pd.isna(x)) or ( x == "nan")) :
        return x
    elif type(x) == float or type(x) == int :
        return x
    elif type(x) == str :
        return x 
    else :
        return x
```

```python
df.consumption = df.consumption.transform(fix).str.split(" l").str[0]
df.consumption = df.consumption.transform(fix).str.split(" kg").str[0]

# kg mın lt eşit olmadığının farkındayız. üşendik çevirmeye (:
```
